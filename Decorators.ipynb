## **Decorators**

Powerfull tool in Python. There are functions (they are the more common type) and class decorators. Above the definition of your function, you put a function decorator name. It uses @ to declare a decorator. 

THe decorator takes another function as a parameter and it allows to have new functionality for an existing function. Function in Python are first class objects. It means  that they can be passed as arguments, returned from a function, modified, etc. 

More info: https://www.datacamp.com/community/tutorials/decorators-python


```python
def start_end_decorator(func):
  #inner function which wraps the call to the function
  def wrapper():
    print('Start')
    func()
    print('End')
  return wrapper

@start_end_decorator
def print_name():
  print('Miguel')


#print_name = start_end_decorator(print_name)

print_name()
```

```python

import functools

def start_end_decorator(func):


  #inner function which wraps the call to the function
  @functools.wraps(func)
  def wrapper(*args, **kwargs):
    print('Start')
    result = func(*args, **kwargs)
    print('End')
    return result
  return wrapper

@start_end_decorator
def add5(x):
  return x + 5

result = add5(10)
print(result)
print(help(add5))
print(add5.__name__)
```

```python
#Decorators with argumens

import functools

def repeat(num_times):
  def decorator_repeat(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        for _ in range(num_times):
          result = func(*args, **kwargs)
        return result
    return wrapper
  return decorator_repeat

@repeat(num_times=3)
def greet(name):
  print(f'Hello {name}')

greet('Alex')
```
