## **Lambda Functions**

Functions that could be used once in your code, or if you want to hide the call of another function

```python
def add10_fun(x):
  return x + 10

add10 = lambda x: x + 10
print(add10(5))

mult = lambda x, y: x*y
print (mult(2, 7))
```

```python
a = [1, 2, 3, 4, 5]
b = map(lambda x: x*2, a)
print(list(b))

c = [x*2 for x in a]
print (c)
```

```python
#filter function
a = [1, 2, 3, 4, 5, 6]
b = filter(lambda x: x%2==0, a)
print (list(b))
```

```python
#reduce function is like the group by in SQL
from functools import reduce

a = [1, 2, 3, 4, 5, 6]
b = reduce(lambda x,y: x*y, a)
print(b)
```
