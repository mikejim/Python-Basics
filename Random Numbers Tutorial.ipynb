## **Random Numbers**

```python
import random

#Pseudo-random

a = random.random()
print(a)

b = random.uniform(1, 10)
print(b)

c = random.randint(1, 10)
d = random.randrange(1, 10) # the top range is not included
e = random.normalvariate(0, 1) # Random value from a Normal distribution
print(c)

mylist = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
a = random.choice(mylist)
b = random.sample(mylist, 3)
print(a)
print(b)

random.shuffle(mylist)
print(mylist)
```

```python

#You can reproduce your numbers by using the same seed
random.seed(1)
print(random.random())
print(random.randint(1,10))

random.seed(2)
print(random.random())
print(random.randint(1,10))
```

**Secrets**

Used for passwords, security tokens, etc.

```python

import secrets

a = secrets.randbelow(10)
print(a)

# It will use 4 bits (1111 means the max number is 15)
b = secrets.randbits(4)
print(b)

mylist = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
c = secrets.choice(mylist)
print(c)
```

**Using NumPy**

```python
import numpy as np
a = np.random.rand(3)
print(a)

# From 0 to 9, a total of 3 elements will be generated
b = np.random.randint(0, 10, 3)
print (b)

# From 0 to 9, an array of [3,4] elements will be generated
b = np.random.randint(0, 10, (3, 4))
print (b)

np.random.shuffle(b)
print(b)

#This is a different seed function than the previous one
np.random.seed(1)
print(np.random.rand(3,3))

np.random.seed(1)
print(np.random.rand(3,3))
```
