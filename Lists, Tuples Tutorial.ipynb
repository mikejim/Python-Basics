**Lists**

```python
#Lists: ordered, mutable, allows duplicate elements

mylist = ["banana", "cherry", "apple"]
print(mylist)

if "lemon" in mylist:
  print("yes")
else:
  print("no")

#Print the length of our list
print(len(mylist))

mylist.append("mango")
print(mylist)

mylist.insert(1, "aguacate")
print(mylist)

#Returns and removes the last item
item = mylist.pop()
print (item)
print(mylist)

#Be careful it might return a compiling error if the element is not found
mylist.remove("cherry")
print(mylist)

#sort list
list2 = sorted(mylist)
print("New list:", list2)

#Remove all elements
mylist.clear()
```

```python
mylist = [0]*5
print(mylist)

mylist1 = [1,2,3,4,5]
mylist2 = mylist + mylist1
print(mylist2)
```

## **Tuples**
ordered, immutable, allows duplicate elements

```python
mytuple = "Max", 25, "Boston"
print(mytuple)
print(type(tuple))

tuple01 = tuple(["Max", 24, "NYC"])
print(tuple01)
print(type(tuple01))

#This is not allowed because is immutable
#tuple01[0] = 23

if "Max" in tuple01:
  print("yes")
else:
    print("No")

print(len(tuple01))

print("Index of element: ", tuple01.index('NYC'))

print(tuple01.count('NYC'))

list01 = list(tuple01)
print(list01)
tuple2 = tuple(list01)
print(tuple2)
```

```python
tuple02 = "Quilla", "Bog", "Cali", "Medellin"

c1, c2, c3, c4 = tuple02
print (c1, c2)

tuple03 = (0,1,2,3,4)
print(type(tuple03))

i1, *i2, i3 = tuple03
print(i1)
# i2 ends up with all the remaining elements
print(i2)
print(i3)
```
