Key-value pairs, unordered, mutable

````python
dict01 = {"name": "Max", "age": 29, "city": "NYC"}
print(dict01)

value = dict01["age"]
print(value)

#You can add or change items
dict01["email"] = "max@gmail.com"
print(dict01)

#You can modify
dict01["email"] = "dog@gmail.com"
print(dict01)

del dict01["name"]
print (dict01)

dict01.pop("email")
print(dict01)

dict01["email"] = "dog@gmail.com"
print(dict01)

if "email" in dict01:
  print("yes it is")

try:
  print(dict01["lastname"])
except:
  print("Error")
````

````python
for key in dict01:
  print(key)

print("--------------------------")
for valu in dict01.values():
  print(valu)

#Be careful with copies
dictcopy = dict01
dictcopy["email"] = "perro@gmail.com"

print (dictcopy)
print(dict01)
````

````python
#You can merge dictionaries
dict01 = {"name": "Miguel", "age": 23, "email":"migue@gmail.com"}
dict02 = {"name": "Ana", "age": 23, "email":"migue@gmail.com", "tel": "301"}
print(dict01)

dict01.update(dict02)
print(dict01)
````

## **Sets**

Unordered, mutable, no duplicates

````python
set01 = set("Hello")
set02 = set()
print(set01)
print(type(set01))

set01.add(1)
set01.add(2)

print(set01)

set01.discard(2)

for i in set01:
  print(i)

if 1 in set01:
  print("yes 1 is in")
````

````python
#Union and intersection

odds = {1, 3, 5, 7, 9}
evens = {0, 2, 4, 6, 8}
primes = {2, 3, 5, 7, 11}

#Combines without duplication
u = odds.union(evens)
print(u)

#Prints an empty set
i = odds.intersection(evens)
print(i)

i = odds.intersection(primes)
print(i)

setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11}
setC = {4, 5, 6}

#Prints those in A but not those found in B
diff = setA.difference(setB)
print(diff)

#Prints the difference between the two sets, it means not those repeated
diffSim = setB.symmetric_difference(setA)
print(diffSim)

setC.update(setA)
print(setC)
````
