## **Strings**

ordered, immutable, text representation

```python
string01 = "I'm a programmer"
print(string01)

string02 = "Hello " \
"World"
print(string02)

string03 = "This is a sentence"
char = string03[0]
#Gets the second last character
char1 = string03[-2]
print(char)
print(char1)

substring = string03[1:5]
print(substring)

#If I don't use a start index, it will go from the beginning
subs01 = string03[:5]
print(subs01)

#It takes every two characters
subs01 = string03[::2]
print(subs01)

#It reverses the string 
subs01 = string03[::-1]
print(subs01)
```

```python
string = "Hello"
name = "Tom"
sentence = string + " " + name
print(sentence)

for i in sentence:
  print(i)

if "e" in sentence:
  print("Yes e is found")

if "Hel" in sentence:
  print("Yes Hel is found")
```

```python
string02 = "      Hello World      x"
print(string02)
string02 = string02.strip()
print(string02)

x = string02.upper()
print(x)

print(string02.startswith("A"))
print(string02.endswith("x"))

print(string02.find("o"))
y = string02.replace("World", "Universe")
print(y)
```

**Printing numbers with strings**

```python
var = 3.12
mystring = "The variable is %f" % var
print(mystring)

var = 3
mystring = "The variable is %d" % var
print(mystring)

var = 3.1415
mystring = "The variable is %.2f" % var
print(mystring)

#NEW METHOD
var = 3.1415
mystring = "The variable is %.2f" % var
print(mystring)

var = 3.1415
var2 = 4
mystring = "The variable is {:.2f} and {}".format(var, var2)
print(mystring)

#This is the new since Python 3.6, f-Strings
#It evaluates at runtime
var = 3.1415
var2 = 4
mystring = f"The variable is {var} and {var2}"
print(mystring)

mystring = f"The variable is {var*2} and {var2}"
print(mystring)


```
